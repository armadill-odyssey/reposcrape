```python
# import pandas as pd

filename = '/Users/coreysery/Developer/School/CSC3130/reposcrape/data/2017-02-09.github.json'
# data = pd.read_json(filename)

import pandas as pd
import json
import os

# Reading the json as a dict
with open(filename) as json_data:
    data = json.load(json_data)
```

```python
data['all']
```

```python
# Top 50 Trending repos with most stars
import operator

results = {}

for lang, repos in data.items():
    for r in repos:
        name = r['repo']
        stars = r['stars']
        results[name] = stars
                
sorted_results = sorted(results.items(), key=operator.itemgetter(1), reverse=True)

top50 = sorted_results[0:50]

for r in top50:
    print('%s: %d' % r)

```
